## AC Stark Experiment

```python
center_est = -30 * 0.65 * MHz
ac_freq_span = 20 * MHz
ac_num_exp = 141
ac_freq_linspace = np.linspace(-0.5 * ac_freq_span + center_est, 0.5 * ac_freq_span + center_est, ac_num_exp)
```

```python
ac_exp_1 = []
ac_exp_2 = []

meas_pulse = pulse.GaussianSquare(
    duration=2 * measure_pulse.duration,
    amp=measure_pulse.amp,
    sigma=measure_pulse.sigma,
    width=2*measure_pulse.duration - 4 * measure_pulse.sigma,
    angle=measure_pulse.angle
)

for freq_shift in ac_freq_linspace:
    with pulse.build(backend=backend, default_alignment='sequential', name=f'AC Freq 1, shift: {int(1e3*freq_shift/MHz)/1e3}MHz') as ac_sched_1:
        meas_chan = pulse.measure_channel(qubit)
        qubit_chan = pulse.drive_channel(qubit)

        pulse.shift_frequency(freq_shift, qubit_chan)

        with pulse.align_right():
            pulse.play(meas_pulse, meas_chan)
            pulse.play(x_pulse, qubit_chan)
            pulse.delay(
                duration=4*measure_pulse.sigma,
                channel=qubit_chan
            )
        
        pulse.delay(measure_delay.duration, meas_chan)
        pulse.measure(qubit, pulse.MemorySlot(qubit))
    ac_exp_1.append(ac_sched_1)

    with pulse.build(backend=backend, default_alignment='sequential', name=f'AC Freq 2, shift: {int(1e3*freq_shift/MHz)/1e3}MHz') as ac_sched_2:
        meas_chan = pulse.measure_channel(qubit)
        qubit_chan = pulse.drive_channel(qubit)

        pulse.shift_frequency(freq_shift, qubit_chan)

        with pulse.align_right():
            pulse.play(meas_pulse, meas_chan)
            pulse.play(x_pulse, qubit_chan)
            pulse.delay(
                duration=6*measure_pulse.sigma,
                channel=qubit_chan
            )
        
        pulse.delay(measure_delay.duration, meas_chan)
        pulse.measure(qubit, pulse.MemorySlot(qubit))
    ac_exp_2.append(ac_sched_2)

ac_full_exp = ac_exp_1 + ac_exp_2

ac_exp_1[-1].draw(
    backend=backend,
    style=stylesheet.IQXDebugging(),
    show_framechange=True,
    show_waveform_info=True,
    show_barrier=True
)
```

## Vary Amplitude and Measure Photon, Assert Linear Photon-Power Law

```python
amps_linspace = np.linspace(0.23, 1., 5)

# For amp 0.43, center freq is approx 25 MHz, corresponding to photon of 39.7
# Assuming Freq = k * Photon = k * u * A**2
# so Freq = k * A**2
measured_freq = -27.0 * MHz
amp_used = 0.43
k1 = measured_freq / amp_used**2
center_freq_linspace = k1 * amps_linspace**2
print(center_freq_linspace)
plt.plot(amps_linspace, center_freq_linspace/MHz, label='center freqs')
plt.axvline(x=0.43, label='normal meas amp', linestyle='dashed', color='grey')
plt.xlabel('Amps (AU)')
plt.ylabel('Predicted Freq (MHz)')
plt.legend()
plt.show()
```

```python
linear_span = 20.0 * MHz
num_exp = 57
```

```python
linear_exps = []

for ind, amp in enumerate(amps_linspace):
    t_linspace = np.linspace(-0.5 * linear_span + center_freq_linspace[ind], 0.5 * linear_span + center_freq_linspace[ind], num_exp)
    
    m_pulse = pulse.GaussianSquare(
        duration=2 * measure_pulse.duration,
        amp=amp,
        sigma=measure_pulse.sigma,
        width=2*measure_pulse.duration - 4 * measure_pulse.sigma,
        angle=measure_pulse.angle
    )

    ac_exp = []

    for freq_shift in t_linspace:
        with pulse.build(backend=backend, default_alignment='sequential', name=f'Lin AC Freq amp: {int(1e2*amp)/1e2}, shift: {int(1e3*freq_shift/MHz)/1e3}MHz') as lin_sched:
            meas_chan = pulse.measure_channel(qubit)
            qubit_chan = pulse.drive_channel(qubit)

            pulse.shift_frequency(freq_shift, qubit_chan)

            with pulse.align_right():
                pulse.play(m_pulse, meas_chan)
                pulse.play(x_pulse, qubit_chan)
                pulse.delay(
                    duration=4*measure_pulse.sigma,
                    channel=qubit_chan
                )
            
            pulse.delay(measure_delay.duration, meas_chan)
            pulse.measure(qubit, pulse.MemorySlot(qubit))
        ac_exp.append(lin_sched)
    
    linear_exps.append(ac_exp)

batched_linear_exp = np.array(linear_exps).reshape(-1).tolist()
print(len(batched_linear_exp))

batched_linear_exp[-57].draw(
    backend=backend,
    style=stylesheet.IQXDebugging(),
    show_framechange=True,
    show_waveform_info=True,
    show_barrier=True
)
```
